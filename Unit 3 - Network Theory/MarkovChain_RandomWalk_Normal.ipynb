```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")

n_steps = 1000
steps   = np.random.randn(n_steps)
walk    = np.cumsum(steps)

plt.figure(figsize=(12, 4))
plt.plot(walk, color="#4a90d9", linewidth=0.8)
plt.axhline(0, color="#ffffff33", linewidth=0.8, linestyle="--")
plt.title(f"Random walk ({n_steps} steps)")
plt.xlabel("Step")
plt.ylabel("Position")
plt.tight_layout()
plt.show()
```
