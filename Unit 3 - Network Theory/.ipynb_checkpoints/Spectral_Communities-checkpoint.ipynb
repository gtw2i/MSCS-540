```python
# ────────────────────────────────────────────────────────────────────────────────
# 0) Imports & Dark‐theme Setup
# ────────────────────────────────────────────────────────────────────────────────
import os
import numpy as np
import tensorflow as tf
import networkx as nx
import matplotlib.pyplot as plt
from tensorflow.keras import Model, Input
from tensorflow.keras.layers import Dense, Lambda
from tensorflow.keras.optimizers import Adam
import seaborn as sns

plt.style.use("dark_background")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
tf.get_logger().setLevel("ERROR")

def generate_community_graph(
    n_nodes: int,
    n_communities: int,
    p_intra: float,
    p_inter: float,
    seed: int = None
) -> nx.Graph:
    """
    Generate a graph with communities using the stochastic block model.

    Parameters
    ----------
    n_nodes : int
        Total number of nodes in the graph.
    n_communities : int
        Number of communities (clusters).
    p_intra : float
        Probability of edges within the same community.
    p_inter : float
        Probability of edges between different communities.
    seed : int, optional
        Random seed for reproducibility.

    Returns
    -------
    G : networkx.Graph
        The generated graph.
    community_labels : list[int]
        A list of length n_nodes giving the community assignment for each node.
    """
    if seed is not None:
        np.random.seed(seed)

    # Split nodes evenly into communities
    sizes = [n_nodes // n_communities] * n_communities
    remainder = n_nodes % n_communities
    for i in range(remainder):
        sizes[i] += 1

    # Build probability matrix
    probs = np.full((n_communities, n_communities), p_inter)
    np.fill_diagonal(probs, p_intra)

    # Generate graph using stochastic block model
    G = nx.stochastic_block_model(sizes, probs, seed=seed)

    # Community labels (to keep track of membership)
    community_labels = []
    for comm_id, size in enumerate(sizes):
        community_labels.extend([comm_id] * size)

    return G, community_labels

def spectral_embedding_manual(A, n_components=2):
    # Degree matrix
    d = np.sum(A, axis=1)
    D = np.diag(d)
    
    L = D - A

    # Eigen-decomposition
    w, v = np.linalg.eigh(L)  # eigh since symmetric
    idx = np.argsort(w)       # sort eigenvalues

    # skip the first eigenvector (associated with eigenvalue 0)
    embedding = v[:, idx[1:n_components+1]]

    return embedding
```

```python
# ────────────────────────────────────────────────────────────────────────────────
# 1) Hyperparameters & Data Preparation
# ────────────────────────────────────────────────────────────────────────────────
#np.random.seed(42)
#tf.random.set_seed(42)

N_AGENTS     = 20       # number of agents
K            = 5       # classes per belief vector
LATENT_DIM   = 3       # bottleneck size (< K)
N_SAMPLES    = 100   # samples per agent
BATCH_SIZE   = 10
EPOCHS       = 100
LR           = 1e-2
DIR_ALPHA    = np.ones(K)

# ground‐truth belief
true_belief = np.random.dirichlet(DIR_ALPHA).astype(np.float32)

# build P_data: shape (N_SAMPLES, N_AGENTS, K)
beliefs_list = []
PERTURB = 1.0
for i in range(N_AGENTS):
    noise = np.random.dirichlet(DIR_ALPHA, size=N_SAMPLES).astype(np.float32)
    mixed = (1 - PERTURB) * true_belief + PERTURB * noise
    beliefs_list.append(mixed)

P_data = np.stack(beliefs_list, axis=1)

# ────────────────────────────────────────────────────────────────────────────────
# 2) Build & Visualize the Communication Graph
# ────────────────────────────────────────────────────────────────────────────────
#G = nx.watts_strogatz_graph(N_AGENTS, k=5, p=0.7)
#G = nx.barabasi_albert_graph(N_AGENTS, m=2, seed=42)
#G = nx.random_regular_graph(d=2, n=N_AGENTS, seed=42)
#G = nx.complete_graph(N_AGENTS)
#G = nx.circulant_graph(N_AGENTS, [1, 2]) 

#"""
G, labels = generate_community_graph(
    n_nodes=N_AGENTS,
    n_communities=2,
    p_intra=0.5,
    p_inter=0.03,
    seed=42
)
#""";

assert nx.is_connected(G)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10,10))
nx.draw_networkx_nodes(G, pos, node_color="#bb0000", edgecolors="white", node_size=400)
nx.draw_networkx_edges(G, pos, alpha=0.6, edge_color="#90e0ef")
nx.draw_networkx_labels(G, pos, {i:f"{i}" for i in G}, font_color="white")
plt.title("Agent Graph", color="white")
plt.axis("off")
plt.show()
```

```python
# ────────────────────────────────────────────────────────────────────────────────
# 3) Agent Modules & Utility Layers
# ────────────────────────────────────────────────────────────────────────────────
class Agent(Model):
    def __init__(self, idx):
        super().__init__(name=f"A{idx}")
        self.encoder = Dense(LATENT_DIM, use_bias=True, name=f"A{idx}_enc")
        self.decoder = Dense(K,          use_bias=True, name=f"A{idx}_dec")

    def call(self, x):
        return self.decoder(self.encoder(x))

def inverse_softmax(p):
    logp = tf.math.log(p + 1e-9)
    mean = tf.reduce_mean(logp, axis=-1, keepdims=True)
    return logp - mean

def softmax_logits(z):
    return tf.nn.softmax(z, axis=-1)

inv_soft = Lambda(inverse_softmax, name="inv_softmax")
soft     = Lambda(softmax_logits,   name="softmax")

# ────────────────────────────────────────────────────────────────────────────────
# 4) Build the SuperModel with Flattened Outputs
# ────────────────────────────────────────────────────────────────────────────────
# 4a) instantiate agents and build
agents = [Agent(i) for i in range(N_AGENTS)]
for ag in agents:
    ag(tf.zeros((1, K)))

# 4b) build functional graph
P_in       = Input(shape=(N_AGENTS, K), name="P_in")
flat_outs  = []

for i in range(N_AGENTS):
    p_i  = P_in[:, i, :]
    z_i  = inv_soft(p_i)
    e_i  = agents[i].encoder(z_i)

    for j in range(N_AGENTS):
        # two‐hop i→j→i
        d_j  = agents[j].decoder(e_i)
        e_j   = agents[j].encoder(d_j)
        d_i  = agents[i].decoder(e_j)
        p_j   = soft(d_i)
        flat_outs.append(p_j)

supermodel = Model(inputs=P_in, outputs=flat_outs, name="SuperModel")

# ────────────────────────────────────────────────────────────────────────────────
# 5) Compile with Flattened Losses, Metrics, and Adjacency‐based Weights
# ────────────────────────────────────────────────────────────────────────────────
def kl_loss(y_true, y_pred):
    diff = tf.math.log(y_true + 1e-9) - tf.math.log(y_pred + 1e-9)
    return tf.reduce_mean(tf.reduce_sum(y_true * diff, axis=-1))

def argmax_acc(y_true, y_pred):
    return tf.reduce_mean(
        tf.cast(tf.equal(
            tf.argmax(y_true, -1), tf.argmax(y_pred, -1)
        ), tf.float32)
    )

num_out = N_AGENTS * N_AGENTS
flat_losses  = [kl_loss     for _ in range(num_out)]
flat_metrics = [argmax_acc  for _ in range(num_out)]

# adjacency matrix row‐normalized
A   = nx.to_numpy_array(G, dtype=np.float32)
deg = A.sum(axis=1, keepdims=True)
#deg[deg == 0] = 1
W_flat = (A / deg).flatten().tolist()

supermodel.compile(
    optimizer=Adam(LR),
    loss=flat_losses,
    loss_weights=W_flat,
    metrics=flat_metrics
)

#supermodel.summary(line_length=100)
```

```python
# ────────────────────────────────────────────────────────────────────────────────
# 6) Build Flattened Y_targets and Fit
# ────────────────────────────────────────────────────────────────────────────────
y_flat = []
for i in range(N_AGENTS):
    p_i = P_data[:, i, :]     # (N_SAMPLES, K)
    for j in range(N_AGENTS):
        y_flat.append(p_i)

history = supermodel.fit(
    x=P_data,
    y=y_flat,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    shuffle=True,
    verbose=1
)
```

```python
# ────────────────────────────────────────────────────────────────────────────────
# 7) Plot Loss and Accuracy Histories
# ────────────────────────────────────────────────────────────────────────────────
# Extract loss and accuracy histories
loss_keys = [k for k in history.history if "loss" in k]
acc_keys  = [k for k in history.history if "argmax_acc" in k]

# Sort keys for consistent plotting
loss_keys.sort()
acc_keys.sort()

# Plot all losses
plt.figure(figsize=(12, 4))
for key in loss_keys:
    plt.plot(history.history[key], label=key)
plt.title("KL Loss per Output")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(loc="upper right", fontsize="small")
plt.grid(True, alpha=0.3)
plt.show()

# Plot all accuracies
plt.figure(figsize=(12, 4))
for key in acc_keys:
    plt.plot(history.history[key], label=key)
plt.title("Argmax Accuracy per Output")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
#plt.legend(loc="lower right", fontsize="small")
plt.grid(True, alpha=0.3)
plt.ylim(0,1)
plt.show()
```
