```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.style.use("dark_background")
```

# Grid graphs

```python
# ── Graph factories ───────────────────────────────────────────────────────────

def make_line(n):
    return nx.path_graph(n)

def make_grid_2d(n):
    return nx.grid_2d_graph(n, n)

def make_grid_3d(n):
    return nx.grid_graph(dim=[n, n, n])

# ── Graph visualization ───────────────────────────────────────────────────────

def draw_line(G, ax):
    pos = {i: (i, 0) for i in G.nodes()}
    nx.draw_networkx(G, pos, ax=ax, node_size=50, node_color='red',
                     edge_color='white', with_labels=False, arrows=False)
    ax.set_title("1D Grid", fontsize=14)
    ax.axis('off')

def draw_grid_2d(G, ax):
    pos = {(x, y): (x, y) for x, y in G.nodes()}
    nx.draw_networkx(G, pos, ax=ax, node_size=30, node_color='royalblue',
                     edge_color='white', with_labels=False, arrows=False)
    ax.set_title("2D Grid", fontsize=14)
    ax.axis('off')

def draw_grid_3d(G, ax):
    nodes = list(G.nodes())
    xs, ys, zs = zip(*nodes)
    ax.scatter(xs, ys, zs, c='limegreen', s=20, alpha=0.6)
    for u, v in G.edges():
        ax.plot([u[0], v[0]], [u[1], v[1]], [u[2], v[2]],
                color='white', alpha=0.5, linewidth=0.5)
    ax.set_title("3D Grid", fontsize=14)
    ax.set_axis_off()

# ── Ball size measurement ─────────────────────────────────────────────────────

def ball_size(G, center, r):
    lengths = nx.single_source_shortest_path_length(G, center, cutoff=r)
    return len(lengths)

def measure_ball_growth(G, center, max_r):
    radii = np.arange(1, max_r + 1)
    sizes = np.array([ball_size(G, center, r) for r in radii])
    return radii, sizes

def fit_dimension(radii, sizes, max_r_fit=None):
    """Fit only up to max_r_fit to avoid boundary effects."""
    if max_r_fit is not None:
        radii = radii[:max_r_fit]
        sizes = sizes[:max_r_fit]
    d, log_c = np.polyfit(np.log(radii), np.log(sizes), 1)
    return d

def get_interior_centers(G, n_centers, boundary_frac=0.25, seed=42):
    """Pick centers far from the boundary via lowest average distance."""
    rng   = np.random.default_rng(seed)
    nodes = list(G.nodes())

    avg_dist = {}
    for n in nodes:
        lengths      = nx.single_source_shortest_path_length(G, n)
        avg_dist[n]  = np.mean(list(lengths.values()))

    cutoff   = np.quantile(list(avg_dist.values()), boundary_frac)
    interior = [n for n, d in avg_dist.items() if d <= cutoff]

    chosen_idx = rng.choice(len(interior), size=min(n_centers, len(interior)), replace=False)
    return [interior[i] for i in chosen_idx]

# ── Combined demo ─────────────────────────────────────────────────────────────

def demo_graph_dimension(graphs, max_r, max_r_fit=None, n_centers=8):
    """
    graphs : list of (G, title, color, draw_fn, is_3d)
    Returns a DataFrame with one row per (graph, seed).
    """
    n_graphs = len(graphs)
    records  = []

    fig = plt.figure(figsize=(7 * n_graphs, 20))
    # pre-create axes: rows 0,1,2 × cols 0..n_graphs-1
    # row 0: graph structure (may need 3d projection)
    axes = {}
    for col, (G, title, color, draw_fn, is_3d) in enumerate(graphs):
        proj = '3d' if is_3d else None
        axes[(0, col)] = fig.add_subplot(3, n_graphs, col + 1,
                                         projection=proj)
        axes[(1, col)] = fig.add_subplot(3, n_graphs, n_graphs + col + 1)
        axes[(2, col)] = fig.add_subplot(3, n_graphs, 2 * n_graphs + col + 1)

    for col, (G, title, color, draw_fn, is_3d) in enumerate(graphs):

        # ── row 0: graph structure ────────────────────────────────────────
        draw_fn(G, axes[(0, col)])

        # ── rows 1 & 2: ball growth ───────────────────────────────────────
        centers = get_interior_centers(G, n_centers)
        all_d   = []

        for center in centers:
            radii, sizes = measure_ball_growth(G, center, max_r)
            d_fit        = fit_dimension(radii[2:], sizes[2:], max_r_fit=max_r_fit)
            all_d.append(d_fit)
            records.append({"graph": title, "seed": str(center), "fitted_d": d_fit})

            axes[(1, col)].plot(radii, sizes, color=color, alpha=0.4, linewidth=1.5)
            axes[(2, col)].plot(np.log(radii), np.log(sizes), color=color, alpha=0.4, linewidth=1.5)

        mean_d = np.mean(all_d)
        std_d  = np.std(all_d)

        # mean fit line on log-log
        all_radii, all_sizes = zip(*[measure_ball_growth(G, c, max_r) for c in centers])
        mean_sizes  = np.mean(all_sizes, axis=0)
        fit_radii   = all_radii[0]
        c_fit       = np.mean(np.log(mean_sizes[2:])) - mean_d * np.mean(np.log(fit_radii[2:]))
        log_r_range = np.linspace(np.log(fit_radii[2]), np.log(fit_radii[-1]), 200)
        axes[(2, col)].plot(log_r_range, mean_d * log_r_range + c_fit,
                            '--', color='white', linewidth=2.5,
                            label=f"slope = {mean_d:.2f} ± {std_d:.2f}")

        # fit cutoff marker
        if max_r_fit is not None:
            axes[(1, col)].axvline(max_r_fit, color='yellow', linestyle=':',
                                   linewidth=1.5, label="fit cutoff")
            axes[(2, col)].axvline(np.log(max_r_fit), color='yellow', linestyle=':',
                                   linewidth=1.5, label="fit cutoff")

        # labels
        axes[(1, col)].set_title(f"{title}\nBall Growth", fontsize=12)
        axes[(1, col)].set_xlabel("Radius $r$")
        axes[(1, col)].set_ylabel("$|B(r)|$")
        axes[(1, col)].legend(fontsize=9)

        axes[(2, col)].set_title(f"{title}\nLog-Log  (slope = $d$)", fontsize=12)
        axes[(2, col)].set_xlabel("$\\log(r)$")
        axes[(2, col)].set_ylabel("$\\log(|B(r)|)$")
        axes[(2, col)].legend(fontsize=9)

    plt.suptitle("Graph Dimension:  $|B(r)| \\sim r^d$", fontsize=18, y=1.01)
    plt.tight_layout()
    plt.show()

    # ── DataFrame summary ─────────────────────────────────────────────────────
    df = pd.DataFrame(records)
    display(df.groupby("graph")["fitted_d"].describe().round(3))

    return df

# ── Run ───────────────────────────────────────────────────────────────────────

G_line = make_line(40)
G_2d   = make_grid_2d(20)
G_3d   = make_grid_3d(15)

graphs = [
    (G_line, "1D Grid", "red",       draw_line,    False),
    (G_2d,   "2D Grid", "royalblue", draw_grid_2d, False),
    (G_3d,   "3D Grid", "limegreen", draw_grid_3d, True),
]

df = demo_graph_dimension(graphs, max_r=10, max_r_fit=10, n_centers=8)
```

# Sphereical graph

```python
def make_spherical_graph(n_points=300, angle_threshold_deg=15, seed=42):
    rng  = np.random.default_rng(seed)
    vecs = rng.normal(size=(n_points, 3))
    vecs /= np.linalg.norm(vecs, axis=1, keepdims=True)
    threshold = np.cos(np.radians(angle_threshold_deg))
    G = nx.Graph()
    G.add_nodes_from(range(n_points))
    for i in range(n_points):
        for j in range(i + 1, n_points):
            if np.dot(vecs[i], vecs[j]) >= threshold:
                G.add_edge(i, j)
    nx.set_node_attributes(G, {i: vecs[i] for i in range(n_points)}, 'pos')
    return G, vecs

def draw_spherical_qt(G, vecs, color='orange'):
    """Opens an interactive Qt window for the 3D sphere."""
    fig = plt.figure(figsize=(8, 8))
    ax  = fig.add_subplot(111, projection='3d')
    xs, ys, zs = vecs[:, 0], vecs[:, 1], vecs[:, 2]
    ax.scatter(xs, ys, zs, c=color, s=30, alpha=0.8)
    for u, v in G.edges():
        ax.plot([vecs[u, 0], vecs[v, 0]],
                [vecs[u, 1], vecs[v, 1]],
                [vecs[u, 2], vecs[v, 2]],
                color='white', alpha=0.4, linewidth=0.4)
    ax.set_xlim(-1, 1); ax.set_ylim(-1, 1); ax.set_zlim(-1, 1)
    ax.set_box_aspect([1, 1, 1])
    ax.set_axis_off()
    ax.set_title("Spherical Graph — drag to rotate", fontsize=13)
    plt.tight_layout()
    plt.show()

def demo_ball_growth(G, color='orange', max_r=8, max_r_fit=4, n_centers=10):
    centers = get_interior_centers(G, n_centers)
    all_d   = []
    records = []

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    for center in centers:
        radii, sizes = measure_ball_growth(G, center, max_r)
        saturate_r   = np.argmax(sizes == G.number_of_nodes())
        if saturate_r > 0 and saturate_r < 3:
            continue
        d_fit = fit_dimension(radii, sizes, max_r_fit=max_r_fit)
        all_d.append(d_fit)
        records.append({"graph": "Spherical", "seed": center, "fitted_d": d_fit})
        axes[0].plot(radii, sizes,                 color=color, alpha=0.4, linewidth=1.5)
        axes[1].plot(np.log(radii), np.log(sizes), color=color, alpha=0.4, linewidth=1.5)

    mean_d = np.mean(all_d)
    std_d  = np.std(all_d)

    all_radii, all_sizes = zip(*[measure_ball_growth(G, c, max_r) for c in centers])
    mean_sizes  = np.mean(all_sizes, axis=0)
    fit_radii   = all_radii[0]
    c_fit       = np.mean(np.log(mean_sizes[2:])) - mean_d * np.mean(np.log(fit_radii[2:]))
    log_r_range = np.linspace(np.log(fit_radii[2]), np.log(fit_radii[-1]), 200)
    axes[1].plot(log_r_range, mean_d * log_r_range + c_fit,
                 '--', color='white', linewidth=2.5,
                 label=f"slope = {mean_d:.2f} ± {std_d:.2f}")

    axes[0].axvline(max_r_fit, color='yellow', linestyle=':', linewidth=1.5, label="fit cutoff")
    axes[1].axvline(np.log(max_r_fit), color='yellow', linestyle=':', linewidth=1.5, label="fit cutoff")

    axes[0].set_title("Ball Growth",            fontsize=13)
    axes[0].set_xlabel("Radius $r$"); axes[0].set_ylabel("$|B(r)|$")
    axes[0].legend(fontsize=9)

    axes[1].set_title("Log-Log  (slope = $d$)", fontsize=13)
    axes[1].set_xlabel("$\\log(r)$"); axes[1].set_ylabel("$\\log(|B(r)|)$")
    axes[1].legend(fontsize=9)

    plt.suptitle(f"Spherical Graph — fitted $d$ = {mean_d:.2f} ± {std_d:.2f}  "
                 f"(expected $d$ = 2)", fontsize=15)
    plt.tight_layout()
    plt.show()

    df = pd.DataFrame(records)
    display(df.groupby("graph")["fitted_d"].describe().round(3))
    return df

# ── Cell A: build graph ───────────────────────────────────────────────────────
G_sphere, vecs = make_spherical_graph(n_points=500, angle_threshold_deg=15, seed=42)
print(f"Nodes: {G_sphere.number_of_nodes()},  Edges: {G_sphere.number_of_edges()}")
print(f"Connected: {nx.is_connected(G_sphere)}")
```

```python
# ── Cell B: interactive 3D sphere (Qt window — drag to rotate) ────────────────
%matplotlib qt
plt.style.use("dark_background")
draw_spherical_qt(G_sphere, vecs, color='orange')
```

```python
# ── Cell C: ball growth + dimension estimate (inline) ─────────────────────────
%matplotlib inline
plt.style.use("dark_background")

max_r = 8

df = demo_ball_growth(G_sphere, color='orange', max_r=max_r, max_r_fit=max_r, n_centers=12)
```

# Community graph

```python
def generate_community_graph(
    n_nodes: int,
    n_communities: int,
    p_intra: float,
    p_inter: float,
    seed: int = None
) -> nx.Graph:
    if seed is not None:
        np.random.seed(seed)

    sizes = [n_nodes // n_communities] * n_communities
    remainder = n_nodes % n_communities
    for i in range(remainder):
        sizes[i] += 1

    probs = np.full((n_communities, n_communities), p_inter)
    np.fill_diagonal(probs, p_intra)

    G = nx.stochastic_block_model(sizes, probs, seed=seed)

    community_labels = []
    for comm_id, size in enumerate(sizes):
        community_labels.extend([comm_id] * size)

    return G, community_labels

def demo_ball_growth(G, node_colors=None, edge_color='white', color='mediumpurple',
                     max_r=8, max_r_fit=4, n_centers=10, title="Graph", pos=None):

    fig, axes = plt.subplots(1, 3, figsize=(21, 7))

    # ── col 1: graph ──────────────────────────────────────────────────────────
    if pos is None:
        pos = nx.spring_layout(G, seed=42, k=1.5)

    nx.draw_networkx_nodes(G, pos, ax=axes[0],
                           node_color=node_colors if node_colors is not None else color,
                           node_size=40, alpha=0.9)
    nx.draw_networkx_edges(G, pos, ax=axes[0],
                           edge_color=edge_color, alpha=0.4, width=0.5)
    axes[0].set_title(title, fontsize=13)
    axes[0].axis('off')

    # ── cols 2 & 3: ball growth ───────────────────────────────────────────────
    centers = get_interior_centers(G, n_centers)
    all_d   = []
    records = []

    for center in centers:
        radii, sizes = measure_ball_growth(G, center, max_r)
        saturate_r   = np.argmax(sizes == G.number_of_nodes())
        if saturate_r > 0 and saturate_r < 3:
            continue
        d_fit = fit_dimension(radii, sizes, max_r_fit=max_r_fit)
        all_d.append(d_fit)
        records.append({"graph": title, "seed": center, "fitted_d": d_fit})
        axes[1].plot(radii, sizes,                 color=color, alpha=0.4, linewidth=1.5)
        axes[2].plot(np.log(radii), np.log(sizes), color=color, alpha=0.4, linewidth=1.5)

    mean_d = np.mean(all_d)
    std_d  = np.std(all_d)

    # mean fit line
    all_radii, all_sizes = zip(*[measure_ball_growth(G, c, max_r) for c in centers])
    mean_sizes  = np.mean(all_sizes, axis=0)
    fit_radii   = all_radii[0]
    c_fit       = np.mean(np.log(mean_sizes[2:])) - mean_d * np.mean(np.log(fit_radii[2:]))
    log_r_range = np.linspace(np.log(fit_radii[2]), np.log(fit_radii[-1]), 200)
    axes[2].plot(log_r_range, mean_d * log_r_range + c_fit,
                 '--', color='white', linewidth=2.5,
                 label=f"slope = {mean_d:.2f} ± {std_d:.2f}")

    axes[1].axvline(max_r_fit, color='yellow', linestyle=':', linewidth=1.5, label="fit cutoff")
    axes[2].axvline(np.log(max_r_fit), color='yellow', linestyle=':', linewidth=1.5, label="fit cutoff")

    axes[1].set_title("Ball Growth",            fontsize=13)
    axes[1].set_xlabel("Radius $r$"); axes[1].set_ylabel("$|B(r)|$")
    axes[1].legend(fontsize=9)

    axes[2].set_title("Log-Log  (slope = $d$)", fontsize=13)
    axes[2].set_xlabel("$\\log(r)$"); axes[2].set_ylabel("$\\log(|B(r)|)$")
    axes[2].legend(fontsize=9)

    plt.suptitle(f"{title} — fitted $d$ = {mean_d:.2f} ± {std_d:.2f}", fontsize=15)
    plt.tight_layout()
    plt.show()

    df = pd.DataFrame(records)
    display(df.groupby("graph")["fitted_d"].describe().round(3))
    return df


# ── Run ───────────────────────────────────────────────────────────────────────

G_comm, labels = generate_community_graph(
    n_nodes       = 200,
    n_communities = 3,
    p_intra       = 0.2,
    p_inter       = 0.01,
    #seed          = 42
)

print(f"Nodes: {G_comm.number_of_nodes()},  Edges: {G_comm.number_of_edges()}")
print(f"Connected: {nx.is_connected(G_comm)}")

community_colors = plt.cm.tab10(np.array(labels) / max(labels))
pos_comm         = nx.spectral_layout(G_comm)

df_comm = demo_ball_growth(
    G_comm,
    node_colors = community_colors,
    color       = 'mediumpurple',
    max_r       = 6,
    max_r_fit   = 3,
    n_centers   = 12,
    title       = "Community Graph",
    pos         = pos_comm
)
```
