# Unit 3 — Graph Theory 101 (NetworkX Visual Demos)

This notebook pairs **short definitions** with **NetworkX code** that builds tiny example graphs and visualizes the idea.

We’ll cover the topics from **1a (Intro + core objects)** and **1b (More definitions)**:
- Graph definition $G=(V,E)$
- Undirected vs directed vs weighted graphs
- Degree and neighborhood
- Adjacency matrix and the walk-count identity $A^n$
- Walks vs paths, cycles
- Distance (shortest path / weighted shortest path)
- Cliques
- Connected graphs, connected components
- Induced subgraphs

> Note: Many of these concepts are *about relationships*, not geometry—how you draw the picture doesn’t change the underlying graph.


```python
# ============================================================
# Setup
# ============================================================
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Make plots a bit bigger by default
plt.rcParams["figure.figsize"] = (7, 5)

def draw_graph(G, pos=None, title=None, with_labels=True, node_size=900, node_color=None,
               edge_color=None, width=2.0, arrows=None, edge_labels=None):
    """Helper to draw a graph cleanly with optional edge labels."""
    if pos is None:
        # spring_layout is fine for small toy graphs
        pos = nx.spring_layout(G, seed=7)
    if arrows is None:
        arrows = G.is_directed()

    plt.figure()
    nx.draw_networkx(
        G, pos=pos,
        with_labels=with_labels,
        node_size=node_size,
        node_color=node_color,
        edge_color=edge_color,
        width=width,
        arrows=arrows,
        arrowsize=18,
        font_weight="bold"
    )
    if edge_labels is not None:
        nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels, font_size=10)

    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()

def adjacency_dataframe(G, nodelist=None, weight=None, dtype=int):
    """Return adjacency matrix as a labeled DataFrame."""
    if nodelist is None:
        nodelist = list(G.nodes())
    A = nx.to_numpy_array(G, nodelist=nodelist, weight=weight, dtype=float)
    if dtype is int:
        A = A.astype(int)
    return pd.DataFrame(A, index=nodelist, columns=nodelist)
```

## 1a. Graphs / networks — the basic object

A (simple) graph is written as:

$$G = (V, E)$$

- $V$ is the set of **vertices** (a.k.a. nodes).
- $E$ is the set of **edges** (relationships), often written as pairs $(v_i, v_j)$.

In a graph model, we usually care about **who is connected to whom** (the relation), not the exact geometry of the drawing.

### Example: a tiny undirected social graph

Below, nodes are people and edges mean “knows / is connected to.”


```python
# Undirected example (think: Facebook friendships)
G_und = nx.Graph()
G_und.add_edges_from([
    ("A", "B"),
    ("A", "C"),
    ("B", "C"),
    ("C", "D"),
    ("D", "E"),
])

draw_graph(G_und, title="Undirected graph: edges have no direction")
```

### Undirected vs directed vs weighted

- **Undirected**: $(u,v)$ and $(v,u)$ are the same relationship.
- **Directed**: edges are ordered pairs $(u \to v)$.
- **Weighted**: edges carry a number (e.g., interaction count, cost, distance).

Let’s build a directed “follow” graph and a weighted “interaction” graph.

```python
# Directed example (think: Twitter follows)
G_dir = nx.DiGraph()
G_dir.add_edges_from([
    ("Alice", "Bob"),
    ("Alice", "Cara"),
    ("Bob", "Cara"),
    ("Cara", "Alice"),
])

draw_graph(G_dir, title="Directed graph: arrows matter")

# Weighted example (think: number of comments i made on j)
G_w = nx.Graph()
G_w.add_edge("A", "B", w=5)
G_w.add_edge("A", "C", w=1)
G_w.add_edge("B", "C", w=3)
G_w.add_edge("C", "D", w=2)

edge_labels = {(u, v): d["w"] for u, v, d in G_w.edges(data=True)}
draw_graph(G_w, title="Weighted undirected graph (edge label = weight)", edge_labels=edge_labels)
```

### Node properties: degree and neighborhood

- **Degree** of a node (undirected): number of incident edges.
- In directed graphs:
  - **out-degree** counts outgoing edges,
  - **in-degree** counts incoming edges.
- **Neighborhood**:
  - Undirected: neighbors of $v$ are all nodes adjacent to $v$.
  - Directed: you may distinguish successors (out-neighbors) and predecessors (in-neighbors).


```python
# Degree in an undirected graph
deg = dict(G_und.degree())
deg
```

```python
# Neighborhood (undirected)
v = "C"
neighbors_C = list(G_und.neighbors(v))
neighbors_C
```

```python
# In/out neighborhoods (directed)
v = "Alice"
out_neighbors = list(G_dir.successors(v))      # where v points to
in_neighbors  = list(G_dir.predecessors(v))    # who points to v

out_neighbors, in_neighbors
```

### Adjacency matrix

Given an ordering of nodes, the **adjacency matrix** $A$ is:

- $A_{ij}=1$ if there is an edge from node $i$ to node $j$ (or an undirected edge between them),
- $A_{ij}=0$ otherwise.

A classic identity (for unweighted graphs) is:

$$ (A^n)_{ij} = \text{# of walks of length } n \text{ from } i \text{ to } j $$

Let’s compute $A$, then $A^2$ and $A^3$, and interpret one entry.

```python
# Build adjacency matrix for the undirected example
nodes = ["A", "B", "C", "D", "E"]  # fixed order so interpretation is stable
A = nx.to_numpy_array(G_und, nodelist=nodes, dtype=int)
A_df = pd.DataFrame(A, index=nodes, columns=nodes)
A_df
```

```python
# Walk counts via matrix powers
A2 = np.linalg.matrix_power(A, 2)
A3 = np.linalg.matrix_power(A, 3)

A2_df = pd.DataFrame(A2, index=nodes, columns=nodes)
A3_df = pd.DataFrame(A3, index=nodes, columns=nodes)

A2_df
```

```python
# Example interpretation:
# (A^2)_{A,D} counts the number of 2-edge walks from A to D.
i, j = "A", "D"
count_2walks = A2_df.loc[i, j]
count_2walks
```

```python
# Brute-force check: count length-n walks by explicit enumeration
def count_walks_length_n(G, start, end, n):
    if n == 0:
        return 1 if start == end else 0
    total = 0
    for nbr in G.neighbors(start):
        total += count_walks_length_n(G, nbr, end, n-1)
    return total

count_walks_length_n(G_und, "A", "D", 2), count_2walks
```

## 1b. More definitions (walks, paths, cycles, distance, clique, connectivity, subgraphs)

### Walks vs paths

- A **walk** is a sequence of vertices $(v_0, v_1, \dots, v_k)$ where each consecutive pair shares an edge.
  - Vertices/edges *may repeat*.
- A **(simple) path** is a walk with **no repeated vertices**.
- The **length** of a walk/path is the number of edges used (here, $k$).

Let’s write small helper checks and test a “walk with repeats” vs a simple path.

```python
def is_walk(G, seq):
    """True iff every consecutive pair in seq is an edge in G."""
    return all(G.has_edge(seq[t], seq[t+1]) or G.has_edge(seq[t+1], seq[t]) for t in range(len(seq)-1))

def is_simple_path(seq):
    """True iff no vertex repeats."""
    return len(set(seq)) == len(seq)

walk1 = ["A", "C", "B", "C", "D"]   # repeats C -> a walk, not a simple path
path1 = ["A", "B", "C", "D"]        # no repeats -> simple path

(is_walk(G_und, walk1), is_simple_path(walk1)), (is_walk(G_und, path1), is_simple_path(path1))
```

We can also visualize the same underlying graph; the “walk vs path” distinction is about the *sequence you traverse*, not about changing the graph itself.

```python
draw_graph(G_und, title="Same graph — walk/path is a traversal concept, not a different graph")
```

### Cycle

A **(simple) cycle** is a path that starts and ends at the same vertex with no other repeats.

Example: $A \to B \to C \to A$.

```python
# Make a small cycle graph and highlight it
G_cycle = nx.Graph()
G_cycle.add_edges_from([("A","B"), ("B","C"), ("C","A"), ("C","D")])

pos = nx.spring_layout(G_cycle, seed=3)

# color the cycle nodes
node_colors = ["tab:orange" if n in {"A","B","C"} else "tab:blue" for n in G_cycle.nodes()]
edge_colors = ["tab:orange" if set(e) in [set(("A","B")), set(("B","C")), set(("C","A"))] else "tab:gray"
               for e in G_cycle.edges()]

draw_graph(G_cycle, pos=pos, title="Cycle example: A-B-C-A (highlighted)", node_color=node_colors, edge_color=edge_colors)
```

### Distance

The **distance** $d(u,v)$ is the length of the **shortest path** from $u$ to $v$.

- In unweighted graphs: distance = minimum number of edges.
- In weighted graphs: distance = minimum total weight (edge weights are treated as costs).


```python
# Unweighted shortest path distance in G_und
u, v = "A", "E"
sp = nx.shortest_path(G_und, u, v)              # one shortest path
d = nx.shortest_path_length(G_und, u, v)        # its length

sp, d
```

```python
# Weighted shortest path distance in the weighted graph G_w (edge attribute 'w' is cost)
u, v = "A", "D"
sp_w = nx.shortest_path(G_w, u, v, weight="w")
d_w = nx.shortest_path_length(G_w, u, v, weight="w")

sp_w, d_w
```

### Clique

A **clique** is a set of nodes where **every pair is adjacent** (all-to-all connected).

A $k$-clique is often denoted $K_k$ (a complete graph on $k$ vertices).

```python
# Build a graph that contains a 4-clique on nodes 1,2,3,4
G_clique = nx.Graph()
clique_nodes = [1, 2, 3, 4]
G_clique.add_nodes_from(clique_nodes + [5, 6])
# add all edges among 1..4
for i in clique_nodes:
    for j in clique_nodes:
        if i < j:
            G_clique.add_edge(i, j)

# attach a couple extra nodes
G_clique.add_edge(4, 5)
G_clique.add_edge(5, 6)

# Find maximal cliques
cliques = list(nx.find_cliques(G_clique))
cliques
```

```python
# Visualize (highlight the 4-clique)
pos = nx.spring_layout(G_clique, seed=11)
node_colors = ["tab:orange" if n in clique_nodes else "tab:blue" for n in G_clique.nodes()]
draw_graph(G_clique, pos=pos, title="Clique example: nodes 1-4 form a 4-clique", node_color=node_colors)
```

### Connected graphs and connected components

- A graph is **connected** if *every pair* of vertices has a path between them.
- If a graph is not connected, it splits into **connected components**:
  - maximal connected subgraphs (you can’t add more nodes without breaking connectivity).

Let’s build a graph with two components and detect them.

```python
G_disc = nx.Graph()
# component 1
G_disc.add_edges_from([("A","B"), ("B","C")])
# component 2
G_disc.add_edges_from([("X","Y"), ("Y","Z")])

nx.is_connected(G_disc)
```

```python
components = list(nx.connected_components(G_disc))
components
```

```python
# Visualize components with different colors
pos = nx.spring_layout(G_disc, seed=5)
comp_map = {}
for idx, comp in enumerate(components):
    for n in comp:
        comp_map[n] = idx

node_colors = ["tab:orange" if comp_map[n] == 0 else "tab:blue" for n in G_disc.nodes()]
draw_graph(G_disc, pos=pos, title="Disconnected graph with 2 connected components", node_color=node_colors)
```

### Induced subgraph

Given a graph $G=(V,E)$ and a subset of vertices $S \subseteq V$, the **induced subgraph** on $S$:
- keeps **exactly** the vertices in $S$
- keeps **all** edges of $G$ whose endpoints are both in $S$

NetworkX makes induced subgraphs easy with `G.subgraph(S)`.

```python
# Start from the original undirected graph and induce on a subset
S = ["A", "B", "C"]  # choose a subset of nodes
H = G_und.subgraph(S).copy()  # induced subgraph

draw_graph(G_und, title="Original graph G_und")
draw_graph(H, title="Induced subgraph on S = {A, B, C}")
```
