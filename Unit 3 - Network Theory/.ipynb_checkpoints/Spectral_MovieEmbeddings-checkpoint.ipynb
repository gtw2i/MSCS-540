# Create data

```python
import numpy as np
import pandas as pd

# ============================================================
# 1) INPUT DATA
# ============================================================
movies_to_principals = {
    "Inception (2010)": ["Christopher Nolan","Leonardo DiCaprio","Joseph Gordon-Levitt","Marion Cotillard","Tom Hardy","Hans Zimmer","Wally Pfister","Lee Smith"],
    "Interstellar (2014)": ["Christopher Nolan","Matthew McConaughey","Anne Hathaway","Jessica Chastain","Michael Caine","Hans Zimmer","Hoyte van Hoytema","Lee Smith"],
    "The Dark Knight (2008)": ["Christopher Nolan","Christian Bale","Heath Ledger","Aaron Eckhart","Gary Oldman","Hans Zimmer","Wally Pfister","Lee Smith"],
    "The Dark Knight Rises (2012)": ["Christopher Nolan","Christian Bale","Tom Hardy","Anne Hathaway","Gary Oldman","Hans Zimmer","Wally Pfister","Lee Smith"],
    "Memento (2000)": ["Christopher Nolan","Guy Pearce","Carrie-Anne Moss","Joe Pantoliano","David Julyan","Wally Pfister","Dody Dorn"],
    "Tenet (2020)": ["Christopher Nolan","John David Washington","Robert Pattinson","Elizabeth Debicki","Kenneth Branagh","Ludwig Göransson","Hoyte van Hoytema","Jennifer Lame"],
    "Dunkirk (2017)": ["Christopher Nolan","Fionn Whitehead","Tom Hardy","Mark Rylance","Kenneth Branagh","Hans Zimmer","Hoyte van Hoytema","Lee Smith"],
    "The Prestige (2006)": ["Christopher Nolan","Hugh Jackman","Christian Bale","Scarlett Johansson","Michael Caine","David Julyan","Wally Pfister","Lee Smith"],
    "Oppenheimer (2023)": ["Christopher Nolan","Cillian Murphy","Emily Blunt","Robert Downey Jr.","Matt Damon","Ludwig Göransson","Hoyte van Hoytema","Jennifer Lame"],

    "Pulp Fiction (1994)": ["Quentin Tarantino","John Travolta","Samuel L. Jackson","Uma Thurman","Bruce Willis","Sally Menke","Andrzej Sekula"],
    "Reservoir Dogs (1992)": ["Quentin Tarantino","Harvey Keitel","Tim Roth","Michael Madsen","Steve Buscemi","Sally Menke","Andrzej Sekula"],
    "Kill Bill: Vol. 1 (2003)": ["Quentin Tarantino","Uma Thurman","Lucy Liu","Vivica A. Fox","David Carradine","RZA","Robert Richardson","Sally Menke"],
    "Kill Bill: Vol. 2 (2004)": ["Quentin Tarantino","Uma Thurman","David Carradine","Michael Madsen","Daryl Hannah","RZA","Robert Richardson","Sally Menke"],
    "Inglourious Basterds (2009)": ["Quentin Tarantino","Brad Pitt","Christoph Waltz","Michael Fassbender","Diane Kruger","Ennio Morricone","Robert Richardson","Sally Menke"],
    "Django Unchained (2012)": ["Quentin Tarantino","Jamie Foxx","Christoph Waltz","Leonardo DiCaprio","Samuel L. Jackson","Ennio Morricone","Robert Richardson","Fred Raskin"],
    "The Hateful Eight (2015)": ["Quentin Tarantino","Samuel L. Jackson","Kurt Russell","Jennifer Jason Leigh","Walton Goggins","Ennio Morricone","Robert Richardson","Fred Raskin"],
    "Once Upon a Time in Hollywood (2019)": ["Quentin Tarantino","Leonardo DiCaprio","Brad Pitt","Margot Robbie","Al Pacino","Robert Richardson","Fred Raskin"],
    
    "Star Wars: A New Hope (1977)": ["George Lucas","Mark Hamill","Harrison Ford","Carrie Fisher","Alec Guinness","John Williams","Gilbert Taylor","Paul Hirsch"],
    "The Empire Strikes Back (1980)": ["Irvin Kershner","Mark Hamill","Harrison Ford","Carrie Fisher","Billy Dee Williams","John Williams","Peter Suschitzky","Paul Hirsch"],
    "Return of the Jedi (1983)": ["Richard Marquand","Mark Hamill","Harrison Ford","Carrie Fisher","Ian McDiarmid","John Williams","Alan Hume","Sean Barton"],
    "Star Wars: The Force Awakens (2015)": ["J.J. Abrams","Daisy Ridley","John Boyega","Adam Driver","Harrison Ford","John Williams","Dan Mindel","Maryann Brandon"],

    "The Lord of the Rings: The Fellowship of the Ring (2001)": ["Peter Jackson","Elijah Wood","Ian McKellen","Viggo Mortensen","Sean Astin","Howard Shore","Andrew Lesnie","John Gilbert"],
    "The Lord of the Rings: The Two Towers (2002)": ["Peter Jackson","Elijah Wood","Ian McKellen","Viggo Mortensen","Orlando Bloom","Howard Shore","Andrew Lesnie","Michael Horton"],
    "The Lord of the Rings: The Return of the King (2003)": ["Peter Jackson","Elijah Wood","Ian McKellen","Viggo Mortensen","Sean Astin","Howard Shore","Andrew Lesnie","Jamie Selkirk"],
    "The Hobbit: An Unexpected Journey (2012)": ["Peter Jackson","Martin Freeman","Ian McKellen","Richard Armitage","Andy Serkis","Howard Shore","Andrew Lesnie","Jabez Olssen"],
}

# ============================================================
# 2) BASIC STRUCTURES
# ============================================================
movies_joined = {
    movie: "\n".join(sorted(people))
    for movie, people in movies_to_principals.items()
}

df = pd.DataFrame(
    {
        "movie": list(movies_joined.keys()),
        "principals_joined": list(movies_joined.values())
    }
).set_index("movie")

df = df["principals_joined"].str.get_dummies(sep="\n")
movies = df.index.to_list()
n_movies = len(movies)

print("=== Movie x Principal Incidence Matrix ===")
display(df)
```

# Calculate movie similarities

```python
# ============================================================
# 3) JACCARD UTILITIES AND MATRIX
# ============================================================
def jaccard(a: set, b: set) -> float:
    union = a | b
    return len(a & b) / len(union) if union else 0.0

people_sets = {movie: set(people) for movie, people in movies_to_principals.items()}

jacc_mat = np.zeros((n_movies, n_movies), dtype=np.float32)

for i, movie_i in enumerate(movies):
    set_i = people_sets[movie_i]
    for j, movie_j in enumerate(movies):
        if j < i:
            jacc_mat[i, j] = jacc_mat[j, i]
        else:
            set_j = people_sets[movie_j]
            jacc_mat[i, j] = jaccard(set_i, set_j)

df_jacc = pd.DataFrame(jacc_mat, index=movies, columns=movies)

print("=== Jaccard (IoU) Matrix ===")
display(df_jacc)

# ============================================================
# 4) BUILD PAIR DF EARLY
# ============================================================
triu = np.triu_indices(n_movies, k=1)

df_pairs = pd.DataFrame({
    "movie_a": [movies[i] for i in triu[0]],
    "movie_b": [movies[j] for j in triu[1]],
    "jaccard_overlap": jacc_mat[triu].astype(float),
})

print("=== Pair DF (initial) ===")
display(
    df_pairs.sort_values(
        ["jaccard_overlap", "movie_a", "movie_b"],
        ascending=[False, True, True]
    ).reset_index(drop=True).head(25)
)
```

# SVD embeddings

```python
from sklearn.metrics.pairwise import cosine_similarity

# ============================================================
# 5) SHARED EMBEDDING DIMENSION
# ============================================================
n_components = 10

# ============================================================
# 6) SVD MOVIE EMBEDDINGS
# ============================================================
X = df.values.astype(float)
U, S, Vt = np.linalg.svd(X, full_matrices=False)

df_svd = pd.DataFrame(
    U[:, :n_components] * S[:n_components],
    index=df.index,
    columns=[f"svd_{i+1}" for i in range(n_components)]
)

print("=== SVD Movie Embeddings ===")
display(df_svd)
print()

df_svd_cosine = pd.DataFrame(
    cosine_similarity(df_svd.values),
    index=df.index,
    columns=df.index
)

print("=== SVD cosine similarities for Inception ===")
display(df_svd_cosine.loc["Inception (2010)"].sort_values(ascending=False))
```

# Spectral embeddings

```python
from sklearn.manifold import SpectralEmbedding

# ============================================================
# 7) SPECTRAL EMBEDDING FROM JACCARD AFFINITY
# ============================================================
df_jacc_affinity = jacc_mat.copy()
np.fill_diagonal(df_jacc_affinity, 1.0)

arr_spectral_jacc = SpectralEmbedding(
    n_components=n_components,
    affinity="precomputed",
    random_state=42
).fit_transform(df_jacc_affinity)

df_spectral_jacc = pd.DataFrame(
    arr_spectral_jacc,
    index=movies,
    columns=[f"spec_jacc_{i+1}" for i in range(n_components)]
)

print("=== Spectral Embedding from Jaccard Affinity ===")
display(df_spectral_jacc)
print()

df_spectral_jacc_cosine = pd.DataFrame(
    cosine_similarity(df_spectral_jacc.values),
    index=df_spectral_jacc.index,
    columns=df.spectral_jacc.index if False else df_spectral_jacc.index
)

print("=== Spectral-Jaccard cosine similarities for Inception ===")
display(df_spectral_jacc_cosine.loc["Inception (2010)"].sort_values(ascending=False))
```

# Comparing performance

```python
# ============================================================
# 8) FINAL PAIR TABLE WITH ALL SIMILARITIES
# ============================================================

# append pairwise SVD cosine to df_pairs
df_pairs["svd_cosine"] = [
    df_svd_cosine.loc[movie_a, movie_b]
    for movie_a, movie_b in zip(df_pairs["movie_a"], df_pairs["movie_b"])
]

# append pairwise spectral cosine to df_pairs
df_pairs["spectral_jacc_cosine"] = [
    df_spectral_jacc_cosine.loc[movie_a, movie_b]
    for movie_a, movie_b in zip(df_pairs["movie_a"], df_pairs["movie_b"])
]

df_pairs = df_pairs.sort_values(
    ["jaccard_overlap", "svd_cosine", "spectral_jacc_cosine", "movie_a", "movie_b"],
    ascending=[False, False, False, True, True]
).reset_index(drop=True)

print("=== Pair DF with Jaccard + SVD cosine + Spectral-Jaccard cosine ===")
display(df_pairs)

print("=== Top pairs with nonzero Jaccard overlap ===")
display(df_pairs.loc[df_pairs["jaccard_overlap"] > 0].head(25))
```
