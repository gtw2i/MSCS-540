```python
import wikipedia as wiki
import random

def random_walk(start_page: str, steps: int):
    current = wiki.page(start_page)
    print(f"{current.title}")
    
    for i in range(steps):
        links = current.links
        next_title = random.choice(links)
        try:
            current = wiki.page(next_title)
            print(f"{current.title}")
        except (wiki.PageError, wiki.DisambiguationError):
            continue
```

```python
n_steps = 50
start_page = "Ice Nine Kills"

random_walk(
    start_page,
    n_steps
)
```
