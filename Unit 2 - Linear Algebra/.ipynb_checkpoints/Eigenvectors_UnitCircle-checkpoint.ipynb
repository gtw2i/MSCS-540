```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use("dark_background")

# Define number of points around the unit circle
n_pts = 100
angles = np.linspace(0, 2 * np.pi, n_pts, endpoint=False)
unit_vectors = np.array([np.cos(angles), np.sin(angles)]).T  # Shape (n_pts, 2)

# Define the transformation matrix A
A = np.array([[1, 0.5],
              [0.5, 1]])

# Compute transformed vectors
transformed_vectors = unit_vectors @ A.T  # Shape (n_pts, 2)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
eigenvectors_scaled = eigenvectors * eigenvalues  # Scale eigenvectors by their eigenvalues

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_aspect('equal', 'box')

# Plot the x and y axes in white
ax.axhline(0, color='white', linewidth=0.5)
ax.axvline(0, color='white', linewidth=0.5)

# Plot static ellipses for both original unit circle and transformed vectors
ax.plot(unit_vectors[:, 0], unit_vectors[:, 1], color='blue', linestyle='--', alpha=0.5, label='Original Circle')
ax.plot(transformed_vectors[:, 0], transformed_vectors[:, 1], color='red', linestyle='--', alpha=0.5, label='Transformed Ellipse')

# Plot column vectors of A
ax.plot([0, A[0, 0]], [0, A[1, 0]], color='green', linestyle='-', linewidth=1.5, label='Column Vector 1 of A')
ax.plot([0, A[0, 1]], [0, A[1, 1]], color='orange', linestyle='-', linewidth=1.5, label='Column Vector 2 of A')

# Plot eigenvectors of A, scaled by their eigenvalues
ax.plot([0, eigenvectors_scaled[0, 0]], [0, eigenvectors_scaled[1, 0]], color='cyan', linestyle='-', linewidth=1.5, label='Eigenvector 1 (scaled)')
ax.plot([0, eigenvectors_scaled[0, 1]], [0, eigenvectors_scaled[1, 1]], color='magenta', linestyle='-', linewidth=1.5, label='Eigenvector 2 (scaled)')

# Initialize the lines for original and transformed vectors for animation
original_line, = ax.plot([], [], color='blue', linewidth=2)
transformed_line, = ax.plot([], [], color='red', linewidth=2)

# Add legend for clarity
ax.legend(loc="upper right")

def init():
    original_line.set_data([], [])
    transformed_line.set_data([], [])
    return original_line, transformed_line

def animate(i):
    # Get the i-th original vector
    x_orig, y_orig = unit_vectors[i]
    # Get the i-th transformed vector
    x_trans, y_trans = transformed_vectors[i]
    # Update the data of the lines
    original_line.set_data([0, x_orig], [0, y_orig])
    transformed_line.set_data([0, x_trans], [0, y_trans])
    return original_line, transformed_line

ani = FuncAnimation(fig, animate, frames=n_pts, init_func=init, blit=True, interval=50)

# To save the animation as a GIF or MP4 file, uncomment the following lines and ensure you have the required packages installed.

ani.save('Eigenvectors_UnitCircle.gif', fps=20)
```
